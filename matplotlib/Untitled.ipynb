### Matplotlib 기본 사용

![img](https://wikidocs.net/images/page/92071/basics_00.png)



Matplotlib 라이브러리를 이용해서 그래프를 그리는 일반적인 방법에 대해 소개합니다.

**Keyword: matplotlib.pyplot, plt.plot(), plt.show(), plt.axis(), 포맷 문자열**

#### 기본 그래프 그리기

* pyplot.plot() 함수에 하나의 숫자 리스트를 입력함으로써 아래와 같은 그래프가 그려집니다.

* plot() 함수는 리스트의 값들이 y 값들이라고 가정하고, x 값 [0, 1, 2, 3]을 자동으로 만들어냅니다.

* matplotlib.pyplot 모듈의 show() 함수는 그래프를 화면에 나타나도록 합니다.

```python
import matplotlib.pyplot as plt
```

```python
plt_plot = plt.plot([4,5,7,10])
plt.show()
```

```python
plt.plot([1,2,3,4],[4,5,7,10])
plt.show()
```

#### x축 y축 label 하기

```python
x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
y = [15, 23, 64, 12, 5, 35, 21]
plt.plot(x, y)
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
plt.show()
```

#### legend 붙이기

```python
x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
y1 = [15, 23, 64, 12, 5, 35, 21]
y2 = [14,12,38,13,6,30,29]
plt.plot(x, y1 , label="Daily")
plt.plot(x, y2 , label="Solid")
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
plt.legend(loc='upper left')
plt.show()
```

#### 제목넣기

```python
x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
y1 = [15, 23, 64, 12, 5, 35, 21]
y2 = [14,12,38,13,6,30,29]
plt.plot(x, y1 , label="Daily")
plt.plot(x, y2 , label="Solid")
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
plt.legend(loc='upper left')
plt.title("Solid & Left")
plt.show()
```

```python
x = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=3, markersize=8, label="First")
plt.plot(x, y2, "ro-", linewidth=3, markersize=8, label="Second")
plt.xlabel("First")
plt.ylabel("Second")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot1_211112")
```
